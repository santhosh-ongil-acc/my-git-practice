# VS Code Setup Guide for Python Development

## What is VS Code?

- Visual Studio Code (VS Code) is a lightweight but powerful source code editor developed by Microsoft.
- VS Code supports debugging, embedded Git control, syntax highlighting, intelligent code completion.



## Why Use VS Code for Python?
- **"Built-in Terminal"**: You can directly **"run Python scripts"**, **"manage virtual environments"**, and execute **"Git command"**s from within VS Code's integrated terminal.
- **"Debugging Capabilities"**: VS Code provides powerful debugging tools for Python, helping you find and fix errors quickly.
- **"Git Integration"**: VS Code has built-in **"Git support"**, simplifying version control and collaboration.
- **"Jupyter Notebook Support"**: You can run and edit Jupyter notebooks directly in VS Code, which is great for data science and interactive programming.
- **"Extensive Extension Support"**: VS Code has a rich ecosystem of extensions, especially for Python, making development more efficient and enjoyable.

- steps to install VS Code on your Windows system:

# OpenRouter
## What is OpenRouter?
- OpenRouter is an open-source platform that acts as a unified API for various large language models (LLMs). It allows developers to access multiple LLMs through a single API endpoint.


## Why use OpenRouter?
- Access to Multiple Models
- Load Balancing and Fallback
- Unified API
- Open Source it offers transparency and community-driven development.

## How does OpenRouter work?
- OpenRouter acts as a proxy between your application and the various underlying language models.

- Your application makes an API request to OpenRouter's endpoint.
- OpenRouter receives the request and decides which language model should handle it based on configurations or availability.
- The request is forwarded to the chosen language model's API.
- The response from the language model is sent back to OpenRouter.
- OpenRouter forwards the response back to your application.

## Get an OpenRouter API key:

- Go to the OpenRouter website.
- Sign up for an account using your email or preferred method.
- Once logged in, navigate to the API Keys section.
- Generate a new API key. Copy and securely store this key, as it will be needed to access OpenRouter's services.



- Select "API provider" as "OpenRouter" and paste the API key obtained from the OpenRouter site.
- Select the model based on the requirements.
